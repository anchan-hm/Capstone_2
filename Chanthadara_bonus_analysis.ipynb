## Filtering, Sorting & Exploration

#### Q1:
Create a list of all transactions from January 2024, sorted by sale amount from highet to lowest. Which transactions had a sale amount greater than `$500`?

```python
# converting Transaction Date to datetime
store_sales["Transaction Date"] = pd.to_datetime(store_sales["Transaction Date"]) # pulling month & year

# filtering for January 2024
jan_2024 = store_sales[(store_sales["Transaction Date"].dt.year == 2024) # checking for year 2024
                    & (store_sales["Transaction Date"].dt.month == 1)] # checking for month Jan

# sorting Janurary transactions from highest to lowest sale amount
jan_sort = jan_2024.sort_values(by="Sale Amount", ascending=False)

# displaying output for highest to lowest
jan_sort
```

Time period starts from January 9, 2024 and ends at January 31, 2024.
They are sorted from highest total revenue at the top, and lowest at the bottom

```python
# filtering Jan list to show transactions over $500 ONLY
jan_above_500 = jan_sort[jan_sort["Sale Amount"] > 500]

# displaying output for transactions over $500
jan_above_500
```

The transactions that are over `$500` start at the highest with `$1,763.70` and the lowest with `$603.50`

#### Q3:
Which stores in your assigned territories had at least one transaction with a sale amount over `$1,000`? How many such transactions did each store have?

## Aggregation & NumPy

#### Q4:
What is the total sales revenue across all in-store transactions in the dataset? What is the average transaction amount? What is the standard deviation?

```python
# calculating total sales revenue per in-store transactions
total_revenue = store_sales["Sale Amount"].sum() # summing sale amounts

# calcualting average transaction amount per in-store transactions
avg_transaction = store_sales["Sale Amount"].mean() # avg sale amounts

# calculating standard deviation per transaction
std_transaction = store_sales["Sale Amount"].std() # standard deviation of sale amount

# displaying output
print(f"Total Revenue Per Store: {total_revenue:.2f}")
print(f"Average Transactions Per Store: {avg_transaction:.2f}")
print(f"Standard Deviation Per Sale: {std_transaction:.2f}")
```

#### Q5:
Which stores had total revenue above the average store revenue across the full dataset?

```python
# calculating total revenue per store
store_totals = store_sales.groupby("Store ID")["Sale Amount"].sum().reset_index() # grouping stores and summing revenue

# calculating average store revenue across stores
avg_store_rev = np.mean(store_totals["Sale Amount"]) # grabbing average revenue

# filtering stores with above average revenue
above_avg_stores = store_totals[store_totals["Sale Amount"] > avg_store_rev] # keeping ONLY stores above avg

# merging store locations
above_avg_stores = above_avg_stores.merge(store_detail[["Store ID","Store Location"]], # adding city names per store
                                          on = "Store ID") # matching on Store ID

# displaying output
print(f"Average Total Revenue: {avg_store_rev:.2f}")
above_avg_stores
```

#### Q6:
What is the total revenue for each product category across all territories, sorted from highest to lowest? Which category has the most individual transactions?

```python
# merging store_sales with products
sales_w_cat = store_sales.merge(products[["Prod Num","CategoryID"]], # pulling products & category ID columns
                                    on = "Prod Num") # matching product number

# merging to pull Category name
sales_w_cat = sales_w_cat.merge(product_categories[["CategoryID","Category"]], # pulling category name
                                on = "CategoryID") # merging on CategoryID

# calculating total revenue per product category
cat_totals = sales_w_cat.groupby("Category")["Sale Amount"].sum().reset_index() # grouping categories and summing revenue

# sorting categories from highest to lowest revenu
cat_totals_sort = cat_totals.sort_values("Sale Amount", ascending=False) # sorting revenue highest to lowest

cat_totals_sort = cat_totals_sort.reset_index(drop=True) # resetting index to print in sorted order

# counting number of transactions per category
cat_transactions = sales_w_cat.groupby("Category")["Sale Amount"].count().reset_index() # counting transactions
cat_transactions_sort = cat_transactions.sort_values("Sale Amount", ascending=False) # highest count at top

# formatting output
print("Total Revenue by Category (Highest to Lowest):") # header for revenu table
# looping through each row to format
for i in range(len(cat_totals_sort)): # looping through each index
    category = cat_totals_sort.loc[i, "Category"] # pulling category name
    amount = cat_totals_sort.loc[i, "Sale Amount"] # pulling revenue amount
    print(category + " - $" + format(amount, ",.2f")) # printing category & format revenue

# displaying output
print("\nCategory with Most Transactions:") 
top_cat = cat_transactions_sort.iloc[0] # top row = most transactions
print(top_cat["Category"] + " - " + format(top_cat["Sale Amount"], ",")) 
```

#### Q7:
Across the full dataset, what are the top 10 best-selling individual products by total revenue?

```python
# merging store_sales with products
sales_w_cat = store_sales.merge(products[["Prod Num","CategoryID","Product"]], # pulling prodct number, category ID, product name
                                on = "Prod Num") # matching product number

# merging to pull Category name
sales_w_cat = sales_w_cat.merge(product_categories[["CategoryID","Category"]], # pulling category name
                                on = "CategoryID") # merging CategoryID

# calculating total revenue per product
prod_totals = sales_w_cat.groupby("Product")["Sale Amount"].sum().reset_index() # grouping product and summing revenue

# sorting products from highest to lowest (revenue)
prod_total_sort = prod_totals.sort_values("Sale Amount", ascending=False) # sorting highest to lowest

# resetting index to sort order print
prod_total_sort = prod_total_sort.reset_index(drop=True)

# setting top 10 products (best-selling)
top_prods = prod_total_sort.head(10) # pulling first 10

# printing format
print("Top 10 Best-Selling Products by Total Revenue:") # header

# looping through each row
for i in range(len(top_prods)): # looping through each index
    product = top_prods.loc[i, "Product"] # pulling product name
    amount = top_prods.loc[i, "Sale Amount"] # pulling revenue
    print(product + " - $" + format(amount, ",.2f")) # printing product formatted
```

#### Q9:
What is total revenue by region across the full dataset? Which region generates the most revenue?

```python
# merging store_sales & store_detail to pull region info
sales_w_region = store_sales.merge(store_detail[["Store ID","Region"]], # pulling region per store
                                   on = "Store ID") # matching store ID

# calculating total revenue per region
region_totals = sales_w_region.groupby("Region")["Sale Amount"].sum().reset_index() # grouping and summing region rev

# sorting regions highest to lowest revenue
region_totals_sort = region_totals.sort_values("Sale Amount", ascending=False) # highest to lowest

# resetting index to order print
region_totals_sort = region_totals_sort.reset_index(drop=True)

# formatting ouput
print("Total Revenue by Region (Highest - Lowest):") # header for output

# looping through each row to print formatted values
for i in range(len(region_totals_sort)): # looping through each index
    region = region_totals_sort.loc[i, "Region"] # pulling region name
    amount = region_totals_sort.loc[i, "Sale Amount"] # pulling revenue amount
    print(region + " - $" +format(amount, ",.2f")) # print region and revenue formatted

# locating region with MOST revenue
top_region = region_totals_sort.loc[0, "Region"] # pulling region after sorting
print("\nRegion with the Most Revenue:", top_region) # dispalying region with highest revenue
```

#### Q10:
For each product subcategory in your territories, calculate the total revenue and number of transactions. Which subcategories are driving the most revenue?

```python
# filtering store_detail for ONLY TX & SC
assigned_stores = store_detail[(store_detail["State"] == "Texas")|
                                (store_detail["State"] == "South Carolina")] # keeping TX & SC

# merging store_sales with assigned_stores to pull selected transactions
territory_sales = store_sales.merge(assigned_stores[["Store ID","Store Location","State",
                                                     "Territory Manager"]],
                                    on = "Store ID") # matching store ID

# merging to pull product name, category ID, subcategory ID
territory_sales = territory_sales.merge(products[["Prod Num","CategoryID","SubcategoryID",
                                                  "Product"]],
                                        on = "Prod Num") # matching product number

# merging product_categories to pull category & subcategory names
territory_sales = territory_sales.merge(product_categories[["CategoryID","SubcategoryID",
                                                            "Category","Subcategory"]],
                                        on = ["CategoryID","SubcategoryID"]) # matching IDs

# calculating total revenue AND number of transactions per subcategory
subcat_totals = territory_sales.groupby("Subcategory").agg({"Sale Amount":
                                                            ["sum","count"] #summing rev & count transactions
                                                           }).reset_index() 

# renaming columns
subcat_totals.columns = ["Subcategory","Total Revenue","Transaction Count"]

# sorting subcategories highest to lowest revenue
subcat_sort = subcat_totals.sort_values("Total Revenue", ascending=False)

# resetting index for cleaning print
subcat_sort = subcat_sort.reset_index(drop=True)

# formatting output
print("Total Revenue & Transaction Count by Subcategory (Highest - Lowest):")

# looping through each row to format values
for i in range(len(subcat_sort)):
    subcat = subcat_sort.loc[i, "Subcategory"] # pulling subcategory name
    revenue = subcat_sort.loc[i, "Total Revenue"] # pulling total revenue
    count = subcat_sort.loc[i, "Transaction Count"] # pulling amount of transactions
    print(subcat + " - Revenue: $" + format(revenue, ",.2f") + " | Transactions: "
    + format(count, ",")) 

# locating subcategory with top revenue
top_subcat = subcat_sort.loc[0, "Subcategory"]
print("\nSubcategory with Most Revenue:", top_subcat)
```

#### Q11:
Using the customer list, identify all rewards members who made purchases in your territories. What is the average transaction value for rewards members compared to non-rewards customers?

```python
# splitting into separate fields
customer_ls_split = customer_list["cust_id|date|time|name|email|phone|sms-opt-out "].str.split("|", expand=True)

# renaming columns
customer_ls_split.columns =["cust_id","date","time","name","email","phone","sms-opt-out"]

# converting cust_id to numeric
customer_ls_split["cust_id"] = pd.to_numeric(customer_ls_split["cust_id"])
```

```python
# filtering store_detail for TX & SC
assigned_stores = store_detail[(store_detail["State"] == "Texas")|
                                (store_detail["State"] == "South Carolina")] # ONLY TX & SC

# merging to keep territory transactions
territory_sales = store_sales.merge(assigned_stores[["Store ID","Store Location","State",
                                                     "Territory Manager"]],
                                    on = "Store ID") # matching store ID

# renaming RewardsID to cust_id to match
territory_sales = territory_sales.rename(columns={"RewardsID":"cust_id"})

# merging customer list to rewards members
territory_sales = territory_sales.merge(customer_ls_split[["cust_id","name","email"]],
                                        on = "cust_id") # matching IDs

# rewards memebers that are not empty
territory_sales["Active_Member"] = territory_sales["name"] == territory_sales["name"]

# calculating average transaction value for rewards members
avg_rewards = territory_sales[territory_sales["Active_Member"] == True]["Sale Amount"].mean()

# calculating average transaction value for rewards non-members
avg_non_rewards = territory_sales[territory_sales["name"] == ""]["Sale Amount"].mean()

# displaying output
print("Average Transaction Value (Rewards Members): $", format(avg_rewards, ",.2f"))
print("Average Transaction Value (Non-Rewards Members): $", format(avg_non_rewards, ",.2f"))
```
